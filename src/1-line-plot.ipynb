### Exercise 1: Line Plot Customization

```python
# TASK: Create a line plot with x values ranging from 0 to 10 and y values as the square of x.
# Customize the plot by adding a title, labels for both axes, and a grid.
import matplotlib.pyplot as plt
import numpy as np

try:
    # Generate data
    x = np.arange(0, 11)
    y = x ** 2

    # Validate input dimensions
    if x.shape != y.shape:
        raise ValueError("Input vectors for x and y must have the same shape.")

    # Create line plot
    plt.figure(figsize=(6, 4))
    plt.plot(x, y)

    # Customizations
    plt.title('Line Plot: y = xÂ²')
    plt.xlabel('X values')
    plt.ylabel('Y values')
    plt.grid(True)

    # Show the plot
    plt.show()
except ImportError as e:
    print(f"Error: Missing library - {e}")
except ValueError as e:
    print(e)
except Exception as e:
    print(f"Runtime error occurred: {e}")
```
