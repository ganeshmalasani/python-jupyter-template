### Exercise 2: Bar Plot with Colors

```python
# TASK: Create a bar plot with the following data: categories = ['A', 'B', 'C', 'D'] and values = [5, 7, 3, 9].
# Use different colors for each bar and add a title to the plot.
import matplotlib.pyplot as plt
import numpy as np

try:
    # Define data
    categories = ['A', 'B', 'C', 'D']
    values = [5, 7, 3, 9]

    # Validate input lengths
    if len(categories) != len(values):
        raise ValueError("Error: Mismatched lengths of categories and values.")

    # Validate positive values
    if any(v < 0 for v in values):
        raise ValueError("Error: All values must be non-negative.")

    # Generate dynamic colors
    cmap = plt.get_cmap("viridis")
    colors = cmap(np.linspace(0.2, 0.8, len(categories)))

    # Create bar chart
    plt.figure(figsize=(6, 4))
    plt.bar(categories, values, color=colors)

    # Customizations
    plt.title('Bar Chart')
    plt.xlabel('Categories')
    plt.ylabel('Values')
    plt.grid(axis='y', linestyle='--', alpha=0.7)

    # Show the plot
    plt.show()
except ImportError as e:
    print(f"Error: Missing library - {e}")
except ValueError as e:
    print(e)
except Exception as e:
    print(f"Runtime error occurred: {e}")

```
