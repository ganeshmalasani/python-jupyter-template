### Exercise 5: Subplot Layout

```python
# TASK: Create a 2x2 subplot layout.
# Plot a line chart in the first subplot, a bar chart in the second, a scatter plot in the third, and a pie chart in the fourth.

import matplotlib.pyplot as plt
import numpy as np

def create_subplot(ax, plot_type):
    if plot_type == 'line':
        x = np.arange(0, 6)
        y = x ** 2
        ax.plot(x, y)
        ax.set_title('Line Chart')
    elif plot_type == 'bar':
        categories = ['A', 'B', 'C', 'D']
        values = [5, 7, 3, 9]
        ax.bar(categories, values)
        ax.set_title('Bar Chart')
    elif plot_type == 'scatter':
        x_scatter = [1, 2, 3, 4, 5]
        y_scatter = [2, 4, 6, 8, 10]
        ax.scatter(x_scatter, y_scatter)
        ax.set_title('Scatter Plot')
    elif plot_type == 'pie':
        labels = ['Python', 'Java', 'C++', 'JavaScript']
        sizes = [40, 25, 20, 15]
        ax.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
        ax.set_title('Pie Chart')

try:
    fig, axes = plt.subplots(2, 2, figsize=(10, 8))
    plot_types = ['line', 'bar', 'scatter', 'pie']

    for ax, plot_type in zip(axes.flat, plot_types):
        create_subplot(ax, plot_type)

    plt.tight_layout()
    plt.show()
except ImportError as e:
    print(f"Error: Missing library - {e}")
except Exception as e:
    print(f"Runtime error occurred: {e}")
```
