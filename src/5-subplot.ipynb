### Exercise 5: Subplot Layout

```python
# TASK: Create a 2x2 subplot layout.
# Plot a line chart in the first subplot, a bar chart in the second, a scatter plot in the third, and a pie chart in the fourth.

import matplotlib.pyplot as plt
import numpy as np

try:
    # Create a figure with a 2x2 subplot layout
    fig, axes = plt.subplots(2, 2, figsize=(10, 8))

    # 1st subplot: Line Chart
    x = np.arange(0, 6)
    y = x ** 2
    axes[0, 0].plot(x, y, marker='o', linestyle='-', color='b')
    axes[0, 0].set_title('Line Chart')
    axes[0, 0].set_xlabel('X values')
    axes[0, 0].set_ylabel('Y values')
    axes[0, 0].grid(True)

    # 2nd subplot: Bar Chart
    categories = ['A', 'B', 'C', 'D']
    values = [5, 7, 3, 9]
    colors = ['red', 'blue', 'green', 'orange']
    if len(categories) != len(values) or len(values) != len(colors):
        raise ValueError("Error: Mismatched lengths of categories, values, and colors.")
    axes[0, 1].bar(categories, values, color=colors)
    axes[0, 1].set_title('Bar Chart')

    # 3rd subplot: Scatter Plot
    x_scatter = [1, 2, 3, 4, 5]
    y_scatter = [2, 4, 6, 8, 10]
    if len(x_scatter) != len(y_scatter):
        raise ValueError("Error: x and y must have the same length for scatter plot.")
    axes[1, 0].scatter(x_scatter, y_scatter, color='purple')
    axes[1, 0].set_title('Scatter Plot')

    # 4th subplot: Pie Chart
    labels = ['Python', 'Java', 'C++', 'JavaScript']
    sizes = [40, 25, 20, 15]
    if sum(sizes) != 100:
        raise ValueError("Error: The sum of sizes must be 100 for a percentage-based pie chart.")
    axes[1, 1].pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140, colors=['blue', 'red', 'green', 'orange'])
    axes[1, 1].set_title('Pie Chart')

    # Adjust layout
    plt.tight_layout()
    plt.show()
except ImportError as e:
    print(f"Error: Missing library - {e}")
except ValueError as e:
    print(e)
except Exception as e:
    print(f"Runtime error occurred: {e}")

```
